```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date
import jugaad_data
```

```python
from jugaad_data.nse import stock_df
ticker= ['SBIN','ONGC','CIPLA','JIOFIN','RELIANCE','TCS','ASIANPAINT','INFIBEAM','COALINDIA','WIPRO','ADANIENT','LT','HDFCLIFE','LICI','TRIDENT','ITC','HDFCBANK']
sl = []
for i in ticker:
    df =stock_df(symbol=i, from_date=date(2011,1,1),
            to_date=date(2024,1,1), series="EQ")
    sl.append(df)
```

```python
len(sl)
```

```python
stocks = pd.concat(sl)
stocks
```

```python
stocks.set_index("DATE")
j = 0
for i in stocks.duplicated():
    j+=1
    if(i):
        print(j)
    
```

```python
stocks = stocks.drop_duplicates()
```

```python
stocks.pivot(index="DATE",columns="SYMBOL")
```

```python
close = stocks.loc[:,["CLOSE","DATE","SYMBOL"]].copy()
close = close.pivot(index="DATE",columns="SYMBOL")
close
```

```python
normclose = close.div(close.iloc[0]).mul(100).copy()
normclose.info()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
close.plot(figsize=(10,8),fontsize=10)
plt.show()
```

```python
normclose.plot(figsize=(20,18),fontsize=10)
plt.show()
```

```python
ret = close.pct_change(periods=1)
ret
```

```python
summary = ret.describe().T.loc[:,["mean","std"]]
summary["mean"] = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std",y="mean",figsize=(30,28),s=50,fontsize = 15)
for i in summary.index:
    plt.annotate(i,xy = (summary.loc[i,"std"]+0.0002,summary.loc[i,"mean"]+0.0002),size=10)
plt.xlabel("Annual Risk(std)",fontsize = 15)
plt.ylabel("Annual Return",fontsize = 15)
plt.show()
```
